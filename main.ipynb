## TASK 3

#### Here I created a new DataFrame which is copy of the orignal DataFrame to fill the Nan Values of the price_string column in order to implement my conditional algorithms

```python
df_new = data.copy()
df_new.fillna(-1, inplace=True)
```

#### Here i have added a new Column in the DataFrame which contains a boolean depicting whether the value of the price_string needs to be edited or not , if the value is Nan then False if the first character is dollar than False else the value is True

```python
data.loc[(df_new.price_string !=  -1) | (df_new.price_string.str[0] != '$'), ['have_to_change']] = True
data.loc[(df_new.price_string ==  -1) | (df_new.price_string.str[0] == '$'), ['have_to_change']] = False
```

#### Here I have added the dollar sign in front of those values where have_to_change column which is created above conatins true

```python
data.loc[(data.have_to_change == True), ['price_string']] = '$' + data[data.have_to_change == True].price_string.astype(str)
```

#### Here i have removed our have_to_change column

```python
data.drop(['have_to_change'], axis = 1, inplace=True)
```

#### Here i have added two more column conatining value and currency as asked

```python
data.loc[(df_new.price_string != -1), ['currency']] = data[df_new.price_string != -1].price_string.str[0]
data.loc[(df_new.price_string != -1), ['value']] = data[df_new.price_string != -1].price_string.str[1:]
```

#### Here i changed the data type of the value column from str to float

```python
data.loc[(df_new.price_string != -1), ['value']] = data[df_new.price_string != -1].value.astype(float)
```
